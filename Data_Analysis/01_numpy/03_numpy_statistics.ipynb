### 통계관련 numpy

```python
import numpy as np
```

중심 경향 측도

```python
height = np.array([160, 170, 180, 150, 150, 155, 165, 175, 175, 185, 180, 175, 170, 173, 164, 189])
```

```python
# 평균값
print(np.mean(height))
```

```python
# 중위값
print(np.sort(height))
print(np.median(height))
```

```python
# 최빈값 : numpy로 해본 버전
height = np.array([160, 170, 180, 150, 150, 155, 165, 175, 175, 185, 180, 175, 170, 173, 164, 189])
values, count = np.unique(height, return_counts=True) # .unique() 중복된 값 한개만 가져오는
print(values, count) # [150 155 160 164 165 170 173 175 180 185 189] [2 1 1 1 1 2 1 3 2 1 1]

max_count = np.max(count)
print(count == max_count) # [False False False False False False False  True False False False]
print(values[count == max_count]) # [175]
```

---
### 산포 측도

```python
# 분산 variance
# (값 - 평균)^2의 평균
var1 = np.mean(np.square(height - np.mean(height))) # .square 제곱
var2 = np.var(height)

print(var1) # 128.4375
print(var2) # 128.4375
```

```python
# 표준편차 standard deviation
# 분산의 제곱근
std1 = np.sqrt(var1)
std2 = np.std(height)

print(std1) # 11.333026956643137
print(std2) # 11.333026956643137
```

```python
# 범위
print(np.max(height) - np.min(height)) # 39
```

### 확률 분포

```python
!pip install matplotlib

import matplotlib.pyplot as plt
```

```python
# 정규 분포 
# np.random.randn(shape)
# - 평균이 0, 표준편차가 1인 난수 출력 함수
arr = np.random.randn(1000)

plt.hist(arr, bins=100)    # 히스토그램
plt.show()       # 출력
```

```python
# 정규분포
# np.random.normal(loc, scale, size)
# - loc : 평균
# - scale : 표준편차
# - size : 갯수 or shape
arr = np.random.normal(10, 1, 100)
print(arr)

plt.hist(arr, bins=100)
plt.show()
```

```python
# 균등 분포
# np.random.rand(num)
# - 0.0 이상 ~ 1.0 미만 수에서 균등하게 난수 추출
# - num : 갯수 or shape

arr = np.random.rand(1000)
arr = np.random.rand(2, 3)
print(arr)

plt.hist(arr, bins=100)
plt.show()
```

```python
# 균등 분포
# np.random.randint(start, end, size)
# - start 이상 end 미만의 난수(정수)를 size(shape)만큼 추출
np.random.seed(0)  # 난수 계산 시작값 seed 고정

arr = np.random.randint(1, 100, 1000)
print(arr)

plt.hist(arr, bins=100)
plt.show()
```

```python
arr = np.random.uniform(1, 100, 1000)
plt.hist(arr, bins=100)
plt.show()
```
