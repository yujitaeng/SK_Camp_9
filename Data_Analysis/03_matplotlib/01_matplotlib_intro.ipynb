#### 기본 사용법

```python
x = [1, 2, 3]
y = [4, 5, 6]
plt.plot(x, y)
plt.show()
```

---
### Figure | Axes
- Figure: 전체 그림이 그려지는 컨테이너
- Axes: 그래프가 실제로 그려지는 영역(좌표, 데이터)

```python
fig = plt.figure(figsize=(6, 4), facecolor='green')
print(fig)

ax = plt.gca()
ax.plot([1, 2, 3], [4, 5, 6])
print(ax)

plt.title('Hello Matplotlib')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(10, 6))
print(fig)
print(ax)
plt.show()
```

#### plot 메서드의 x, y 인자

```python
plt.figure(figsize=(3, 2))
```

```python
# 1. 파이썬의 list
```

```python
# 2. ndarray
x = np.arange(1, 11)
y = x * 2
print(x)
print(y)
```

```python
# 3. Series
x = pd.Series(x)
y = pd.Series(y)

print(x)
print(y)
```

```python
# 4. DataFrame
x = pd.DataFrame({'x': x})
y = pd.DataFrame({'y': y})
print(x)
print(y)
```

```python
plt.plot(x, y)
plt.show()
```

### plot 메서드의 매개변수

```python
plt.figure(figsize=(6, 4))


plt.plot(
    x,
    y,
    color='black',
    marker='o',
    markersize=8,
    linestyle='--',
    linewidth=2
)
plt.show()
```

### plt 속성

```python
# label: 기본적으로 한글 폰트 지원하지 않음!
plt.plot(x, y)
plt.title('제목')
plt.xlabel('시간')
plt.ylabel('점수')
plt.show()
```

```python
# ticks 눈금
plt.figure(figsize=(3, 2))

x = np.arange(1, 101)
y = np.linspace(2, 200, 100)

plt.xticks(ticks=np.arange(0, 101, 5), rotation=90)
plt.yticks(rotation=45)

plt.plot(x, y)
plt.show()
```

```python
# lim: 보여지는 범위 재지정
plt.figure(figsize=(3, 2))

x = np.arange(1, 101)
y = np.linspace(2, 200, 100)

plt.xlim(1, 500)
plt.ylim(2, 1000)

plt.plot(x, y)
plt.show()
```

```python
# legend (범례)
plt.figure(figsize=(3, 2))

x = np.arange(1, 101)
y = np.linspace(2, 200, 100)

plt.plot(x, y, label='시간')

plt.legend()

plt.show()
```
