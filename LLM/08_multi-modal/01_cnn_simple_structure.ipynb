# CNN
- tensorflow

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.models import Model
```

```python
# 입력층
# CNN의 입력 차원 = 4차원 = (batch, row, col, channel)
# 배치 차원 생략
input = Input(shape=(28, 28, 1))

# 합성곱층
x = Conv2D(
    filters=16,    # 적용할 필터 개수
    kernel_size=3, # 필터의 크기 (3 x 3)
    strides=1,     # 필터 이동 간격 (오른쪽으로 1, 아래로 1씩 이동)
    padding='valid',
    activation='relu'
)(input)

# 풀링층
x = MaxPooling2D(pool_size=2)(x)

model = Model(input, x)
model.summary()
```

```python
# CNN = 특성 추출 레이어 + 분류 레이어

# 특성 추출 레이어
input = Input(shape=(28, 28, 1))
x = Conv2D(filters=32, kernel_size=3, strides=1, padding='same', activation='relu')(input)
x = Conv2D(filters=64, kernel_size=3, strides=1, padding='same', activation='relu')(x)
x = MaxPooling2D(pool_size=2)(x)

# 분류 레이어
x = Flatten()(x)
x = Dense(100, activation='relu')(x)
x = Dense(30, activation='relu')(x)
output = Dense(10, activation='softmax')(x)

model = Model(input, output)
model.summary()
```
