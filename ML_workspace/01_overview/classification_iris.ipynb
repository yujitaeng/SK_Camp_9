## 최근접 이웃 분류 모델로 훈련/평가 해보기

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
```

```python
# # 데이터 전처리 -> 내 풀이 (라벨데이터에서 오류...)
# # 1. 입력-라벨 데이터
# train_input = iris_data.feature_names
# train_label = iris_data.target_names

# # 2. 훈련-테스트 데이터
# train_input, test_input, train_label, test_label = train_test_split(train_input, train_label, test_size=0.2, random_state=42)

# # 3. 정규화
# standard_scaler = StandardScaler()
# standard_scaler.fit(train_input)
# train_scaled = standard_scaler.transform(train_input)
# test_scaled = standard_scaler.transform(test_input)
```

```python
# 데이터 전처리 -> 선생님 풀이이
# 1. 입력-라벨 데이터 (O)
# 이미 위에서 했음. 고로 패스

# 2. 훈련-테스트 데이터
from sklearn.model_selection import train_test_split
train_input, test_input, train_label, test_label = train_test_split(iris_data.data, iris_data.target, test_size=.25, random_state=42)

train_input.shape, test_input.shape, train_label.shape, test_label.shape


# 3. 정규화
scaler = StandardScaler()
scaler.fit(train_input)
train_scaled = scaler.transform(train_input)
test_scaled = scaler.transform(test_input)

```

```python
# 훈련
kn = KNeighborsClassifier(n_neighbors=7)
kn.fit(train_scaled, train_label)
```

```python
# 평가
kn.score(test_scaled, test_label)
```

```python
# 평가 : accuracy_score
from sklearn.metrics import accuracy_score

pred = kn.predict(test_scaled)

accuracy_score(test_label, pred)
```
